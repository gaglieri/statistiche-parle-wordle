```python
filepath = "/Users/gaglieri/Documents/Programmazione/PYTHON/Analisi-Parle/File-Dizionario-Italiano/italia-1a/ITALIANO_CINQUE_LETTERE.csv"
```

```python
dataframe = pd.read_csv(filepath, header=None, names=["word", "first_letter", "second_letter", "third_letter", "fourth_letter", "fifth_letter"])
```

```python
def make_letter_finder(letter):
    def contains_letter(word): 
        if (letter in word): 
            return True
        else: 
            return False
    return contains_letter
```

```python
def count_words_with_letter(letter):
    finder = make_letter_finder(letter)
    words_with_letter = dataframe.word.map(finder)
    return dataframe.word[words_with_letter==True].count()
```

```python
alphabet = list(map(chr, range(ord('a'),ord('z')+1)))
```

```python
counts_words_with_letter = list(map(count_words_with_letter, alphabet))
```

```python
words_with_letter = pd.DataFrame(list(zip(alphabet, counts_words_with_letter)), columns=["Lettera", "Parole"])
words_with_letter.sort_values("Parole", ascending=False)
```

```python
def make_letter_counter(letter):
    def count_letter_in(word):
        count = 0
        for character in word:
            if (character == letter): count += 1
        return count
    return count_letter_in
```

```python
def get_letter_frequency(letter):
    counter = make_letter_counter(letter)
    letter_counts = dataframe.word.map(counter)
    return letter_counts.sum()
```

```python
letter_frequencies_list = list(map(get_letter_frequency, alphabet))
```

```python
letter_frequencies = pd.DataFrame(list(zip(alphabet, letter_frequencies_list)), columns=["Lettera", "Frequenza"])
letter_frequencies.sort_values("Frequenza", ascending=False)
```
